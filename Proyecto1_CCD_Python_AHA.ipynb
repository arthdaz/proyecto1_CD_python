# Ejercicio No 3
Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:

Captura de pantalla 2022-10-28 222837.png


A continuación, generar nuevas columna:

1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

```python
import pandas as pd
import numpy as np

 
data = {'mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo '],
        'ventas': [30500, 35600, 28300, 33900, 42500],
        'gastos': [22000, 23450, 18100, 35700, 32450]}

#crear dataframe
df_marks = pd.DataFrame(data)
print('DataFrame Original\n------------------')

#agregar columna1
df_marks = (
    df_marks
    .assign(columna1 = lambda x: x.gastos/x.ventas*100  )
)

#agregar columna2
df_marks["columna2"] = df_marks.apply(lambda x: "Meta Superada" if int(x["ventas"]) >= 30000 else "Meta no Superada", axis=1)

#agregar columna3
df_marks["columna3"] = df_marks.apply(lambda x: x.gastos - x.ventas if int(x["gastos"]) > int(x["ventas"]) else 0, axis=1)

df_marks
```

```python
#Total de Ventas ultimos 3 meses
meses = df_marks["mes"]
nmp = meses.to_numpy()
condition = df_marks["mes"].isin(nmp[2:])
df_gastos = df_marks[condition]
df_gastos["ventas"].sum()
```

```python
#Total de gastos en Febrero - Abril
filtro = ["Febrero", "Abril"]
condition = df_marks["mes"].isin(filtro)
df_gastos = df_marks[condition]
df_gastos["gastos"].sum()
```
